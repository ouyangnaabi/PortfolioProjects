```python
#Portfolio Project 4 - Python 
#Find the correlations between the revenue of the film and the other factors
#Import Librarys

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#Read in the data, create a data frame

df = pd.read_csv(r'C:\Users\Kelvin\Desktop\movies.csv')







```

```python
#Let's look at the data

df.head()
```

```python
#Let's see if there's any missing data

for col in df.columns:
    percent_missing = np.mean(df[col].isnull()) #find out the empty columns 
    print('{}-{}%'.format(col, percent_missing)) #return the missing percent in this format
    
# > 0.0% means there are some col that has null values
    




```

```python
#Data types for columns

df.dtypes
```

```python
#Create correct year column
df['Year_correct'] = df['released'].astype(str).str[-20:] #pull the year from the release date as 'year_correct'

df
```

```python
#Order the data by their gross revenue

df = df.sort_values(by=['gross'], inplace = False, ascending = False) #order at desc order 
```

```python
pd.set_option('display.max_rows', None) #Allows you to run the data and see all rows and cols, None = no row limit
```

```python
#Look for duplicates and drop them

#[any column], sort_values(order), shows us all the unique values
df['company'].drop_duplicates().sort_values(ascending = False)
df.drop_duplicates() # this will drop all the duplicates 
```

```python
# What things are the most correlated to the gross revenue
#budget - High Correlation !
#company - prediction

df
```

```python
# Scatter plot with budget vs gross earnings

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget VS Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using Seaborn 
 #shows a positive correlation
    
sns.regplot(x = 'budget', y = 'gross' , data = df, scatter_kws = {"color" : "red"}, line_kws = {"color" : "blue"})
```

```python
# Start looking the correlation
# Pearson, kendall, spearman

#df.corr() #numerical column only
df.corr(method = 'pearson') #budget - gross = 0.74
```

```python
# Visualisation for the correlation table above using heatmap

correlation_matrix = df.corr(method = 'pearson') # create the correlation w/ pearson method

sns.heatmap(correlation_matrix, annot = True) # visualise the correlation

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Look at company 

df.head()
```

```python
# Numerise the company data

df_numerised = df

for col_name in df_numerised.columns:
    if(df_numerised[col_name].dtype == 'object'):
        df_numerised[col_name] = df_numerised[col_name].astype('category') #Change the type
        df_numerised[col_name] = df_numerised[col_name].cat.codes 
        
df_numerised
        
```

```python
df
```

```python
# Create a heatmap for the numerised data above

correlation_matrix = df_numerised.corr(method = 'pearson') # create the correlation w/ pearson method

sns.heatmap(correlation_matrix, annot = True) # visualise the correlation

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Organise to see the highest correlation quickly (unstack it)

df_numerised.corr()
```

```python
correlation_mat = df_numerised.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Show the correlations in a horizontal way (budget genre -0.35...)
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Find all pairs that are higher than 0.5 - the first 2 pairs are the key !

high_corr = sorted_pairs[(sorted_pairs)> 0.5] 

high_corr
```
